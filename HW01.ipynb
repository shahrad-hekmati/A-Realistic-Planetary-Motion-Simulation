#1(الف)- تفاوت های بین HSV, RGB, cmyk

### RGB (قرمز، سبز، آبی) یک مدل افزودنی است که عمدتاً در نمایشگرهای دیجیتال استفاده می شود، که در آن رنگ ها با ترکیب این سه رنگ در شدت های مختلف، از 0 تا 255 ایجاد می شوند. HSV (رنگ، ​​اشباع، ارزش) مدلی است که تراز می کند. بیشتر با ادراک انسان، ارائه یک راه بصری برای انتخاب و تنظیم رنگ ها با نشان دادن آنها از طریق رنگ، اشباع و روشنایی آنها. در مقابل، CMYK (فیروزه‌ای، سرخابی، زرد، کلید/مشکی) یک مدل تفریقی است که در چاپ استفاده می‌شود، که در آن رنگ‌ها با کم کردن درصدهای مختلف نور از پس‌زمینه سفید با استفاده از جوهر تولید می‌شوند. در حالی که RGB و HSV برای برنامه های دیجیتال ضروری هستند، CMYK برای دستیابی به رنگ های دقیق در رسانه های چاپی بسیار مهم است.

*   RGB: بهترین استفاده در نمایشگرهای دیجیتال و برنامه های روشنایی، جایی که امکان انتشار نور وجود دارد. این برای گرافیک ها و برنامه های مبتنی بر وب اساسی است.
*   HSV: اغلب در نرم افزارهای گرافیکی و طراحی استفاده می شود که در آن کاربران نیاز به دستکاری و انتخاب رنگ ها به طور مستقیم دارند. در کارهایی مانند درجه بندی رنگ یا تنظیم تصاویر کمک می کند.

* CMYK: برای چاپ حرفه ای و هر طرحی که بر روی رسانه فیزیکی چاپ می شود ضروری است و اطمینان می دهد که رنگ ها در هنگام چاپ تا حد امکان به طرح مورد نظر نزدیک می شوند.
